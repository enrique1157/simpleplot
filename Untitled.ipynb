```python
#Asignamos nombres a las carpetas de los paises, para que puedan ser leidas y son las siguientes: paises,rios y ciudades.
import os, geopandas as gpd
countries=gpd.read_file(os.path.join("maps","World_Countries","World_Countries.shp"))
rivers=gpd.read_file(os.path.join("maps","World_Hydrography","World_Hydrography.shp"))
cities=gpd.read_file(os.path.join("maps","World_Cities","World_Cities.shp"))
```

```python
#Asigno un nombre para el mapa del pais seleccionado, en este caso Venezuela
venezuela=countries[countries.COUNTRY=='Venezuela']
```

```python
venezuela.plot()
```

```python
#Proyectamos las ciudades de Venezuela
cities_venezuela=cities[cities.COUNTRY=='Venezuela']
cities_venezuela.plot()
```

```python
#Cabecera de las ciudades de Venezuela 
cities_venezuela.head()
```

```python
#Para los rios de Venezuela
rivers_venezuela = gpd.clip(gdf=rivers,
                               mask=venezuela)
rivers_venezuela.plot()
```

```python
#Cabecera de los rios de Venezuela
rivers_venezuela.head()
```

```python
#Version recortada
base = venezuela.plot(facecolor="violet", edgecolor='black', linewidth=0.4,figsize=(5,5))
cities_venezuela.plot(marker='+', color='red', markersize=15,ax=base)
rivers_venezuela.plot(edgecolor='blue', linewidth=0.5,ax=base)
```
